#### Agrupamos el DataFrame 'income_spending' usando el algoritmo KMeans

Usamos un agrupamiento `KMeans` en el DataFrame `income_spending` para encontrar 2 clusters utilizando el algoritmo KMeans, y luego aadimos las etiquetas de los clusters a los datos.

- `modelo_kmeans = KMeans(n_clusters=2)`: Esta lnea crea una instancia del modelo KMeans con 2 clusters. El nmero de clusters es un parmetro que se debe especificar.

- `modelo_kmeans.fit(income_spending)`: Esta lnea ajusta el modelo a los datos de `income_spending`. El modelo KMeans intentar encontrar los centros de los clusters que minimizan la varianza dentro de cada cluster.

- `income_spending['Cluster'] = modelo_kmeans.labels_`: Despus de ajustar el modelo, puedes obtener las etiquetas de los clusters para cada punto de datos utilizando el atributo `labels_` del modelo. Esta lnea aade las etiquetas de los clusters como una nueva columna en el DataFrame `income_spending`.

- `income_spending.head(10)`: Esta lnea muestra las primeras 10 filas del DataFrame `income_spending`, que ahora incluye la columna 'Cluster'.

```python
# Instanciamos el modelo KMeans
modelo_kmeans = KMeans(n_clusters=2)

# Ajustamos el modelo a los datos
modelo_kmeans.fit(income_spending)

# Duplicamos la base de datos 'income_spending' para no modificar la original y agregar los clusters
income_spending_cluster = income_spending.copy()

# Obtenemos las etiquetas de los clusters asignados a cada cliente
income_spending_cluster['Cluster'] = modelo_kmeans.labels_

# Visualizamos los primeros registros de 'income_spending_cluster'
income_spending_cluster.head(10)
```

Vemos que posiblemente 2 no sea el valor ptimo de $k$ que buscamos. Intentemos encontrar el mejor valor de $k$ utilizando el mtodo del codo. Este mtodo implica ejecutar el algoritmo de agrupamiento KMeans para un rango de valores de $k$ y luego trazar el nmero de clusters contra la suma de las distancias al cuadrado dentro de cada cluster. El "codo" en esta trama, es decir, el punto de inflexin donde la tasa de disminucin se vuelve significativamente menor, puede indicar un buen valor de $k$.

#### Evaluamos el mejor valor de $k$ con el mtodo del codo

El mtodo del codo se usa para determinar el nmero ptimo de grupos (clusters) para el algoritmo KMeans.

1. `ks = range(1, 10)`: Aqu se define un rango de nmeros del 1 al 9. Estos representan los posibles nmeros de clusters que se probarn en el algoritmo KMeans.

2. `inercias = []`: Se crea una lista vaca llamada `inercias` para almacenar los valores de inercia para cada nmero de clusters.

3. Luego, se realiza un bucle sobre cada nmero en el rango `ks`:

   - `model = KMeans(n_clusters=k)`: Se crea una instancia del algoritmo KMeans con `k` clusters. El parmetro `random_state` es utilizado para controlar la aleatoriedad en la inicializacin del algoritmo y es til para asegurar que los resultados sean reproducibles.
   
   - `model.fit(samples)`: Se ajusta el modelo a los datos contenidos en `samples`.
   
   - `inercias.append(model.inertia_)`: Se aade el valor de la inercia del modelo a la lista `inercias`. La inercia es una medida de cun dispersos estn los clusters, y se busca minimizarla.

4. Finalmente, se grafican los valores de inercia en funcin del nmero de clusters:

   - `plt.plot(ks, inercias, marker='+', linewidth=0.5, markersize=5)`: Se crea un grfico de lnea entre `ks` e `inercias`. El argumento `marker='+'` especifica que se deben usar cruces (`+`) como marcadores en el grfico. El argumento `linewidth=0.5` establece el ancho de la lnea del grfico a 0.5, y el argumento `markersize=5` establece el tamao de los marcadores a 5.
   
   - `plt.xlabel('Nmero de clusters, k')` y `plt.ylabel('Inercia')`: Se establecen las etiquetas de los ejes x e y.
   
   - `plt.xticks(ks)`: Se establecen las marcas del eje x para que coincidan con los nmeros de clusters.
   
   - `plt.show()`: Se muestra el grfico.

El punto en el que la lnea del grfico se dobla (como un "codo") suele considerarse un buen indicador del nmero ptimo de clusters.

```python
# Definimos un rango de nmeros del 1 al 9 para representar los posibles nmeros de clusters
ks = range(1, 10)

# Creamos una lista vaca para almacenar los valores de inercia para cada nmero de clusters
inercias = []

# Iteramos sobre cada nmero en el rango ks
for k in ks:
    # Creamos una instancia del algoritmo KMeans con k clusters
    model = KMeans(n_clusters=k)
    
    # Ajustamos el modelo a los datos contenidos en income_spending
    model.fit(income_spending)
    
    # Aadimos el valor de la inercia del modelo a la lista inercias
    inercias.append(model.inertia_)
    
# Graficamos los valores de inercia en funcin del nmero de clusters
plt.plot(ks, inercias, marker='+', linewidth=0.5, markersize=5)

# Establecemos la etiqueta del eje x
plt.xlabel('Nmero de clusters, $k$')

# Establecemos la etiqueta del eje y
plt.ylabel('Inercia')

# Establecemos las marcas del eje x para que coincidan con los nmeros de clusters
plt.xticks(ks)

# Mostramos el grfico
plt.show()
```

#### Ajustamos el modelo con el nuevo valor de $k$ y visualizamos los clusters

Observamos que el punto de inflexin ocurre en $k$ = 5, entonces es razonable seleccionar 5 como el nuevo valor ptimo para $k$.

Ahora volvemos a ejecutar el algoritmo KMeans con $k$ = 5 y observamos cmo cambian los clusters. Esto nos proporciona una segmentacin ms precisa de los clientes basada en sus ingresos anuales y puntuaciones de gasto.

```python
# Instanciamos el modelo KMeans
modelo_kmeans = KMeans(n_clusters=5)

# Ajustamos el modelo a los datos
modelo_kmeans.fit(income_spending)
```

```python

# Obtenemos las etiquetas de los clusters asignados a cada cliente
income_spending['Cluster'] = modelo_kmeans.labels_
income_spending.head(10)
```

```python
# Visualizamos los clusters
sns.pairplot(income_spending, hue='Cluster', palette='viridis')
```

#### Hacemos predicciones con los datos de un cliente nuevo

Finalmente, podemos usar el modelo KMeans que se ajust anteriormente para predecir a qu cluster pertenece un nuevo cliente basndonos en sus caractersticas.

- `cliente_nuevo = [ , ]`: Esta lnea crea una lista que representa los datos de un nuevo cliente. Los valores en la lista corresponden a las caractersticas del cliente (ingresos anuales, puntuacin de gasto).

- `cluster_nuevo = modelo_kmeans.predict([cliente_nuevo])`: Esta lnea utiliza el mtodo `predict` del modelo KMeans para predecir a qu cluster pertenece el nuevo cliente. El mtodo `predict` toma una lista de listas como entrada, donde cada lista interna representa los datos de un cliente. En este caso, solo hay un cliente, por lo que la entrada es `[cliente_nuevo]`.

- `print('El nuevo cliente pertenece al cluster:', cluster_nuevo)`: Esta lnea imprime el cluster al que pertenece el nuevo cliente. El mtodo `predict` devuelve una lista de etiquetas de cluster, una para cada cliente en la entrada. En este caso, solo hay un cliente, por lo que `cluster_nuevo` es una lista con un solo elemento.

```python
# Creamos los datos de un nuevo cliente (ingresos anuales, puntuacin de gasto) para hacer una prediccin
cliente_nuevo = [40, 60]

# Hacemos una prediccin con el modelo KMeans para el nuevo cliente
cluster_nuevo = modelo_kmeans.predict([cliente_nuevo])
print('El nuevo cliente pertenece al cluster:', cluster_nuevo)
```
