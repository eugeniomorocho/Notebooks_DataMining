#### Eliminacin de columnas que no aportan informacin til

La decisin de qu columnas eliminar depende en gran medida del contexto y del objetivo del anlisis. Sin embargo, aqu hay algunas pautas generales que podemos considerar:

**Irrelevancia para el modelado**: Las columnas que contienen un valor nico para cada fila (como un ID) generalmente no aportan informacin til para un modelo de data mining y pueden ser eliminadas.

Vamos a usar la funcin `drop` de pandas para eliminar las columnas con valores nicos. 
`data.drop(...)`: Esta es la funcin de pandas que se utiliza para eliminar columnas o filas de un DataFrame.

Por ejemplo, la columna `'Name'` contiene los nombres de los individuos que han solicitado crditos. En general, los nombres de las personas no proporcionan informacin til para un modelo de aprendizaje automtico, ya que son nicos para cada individuo y no tienen una relacin directa con otras variables.

Adems, codificar columnas con valores nicos, podra resultar en un gran nmero de caractersticas nicas, lo que podra hacer el modelo sea ms complejo y difcil de entrenar.

Por lo tanto, en la mayora de los casos, es seguro eliminar la columna 'Name' del conjunto de datos. Sin embargo, hay excepciones. Por ejemplo, si los nombres tienen algn tipo de informacin codificada (como un prefijo que indica el gnero o el estatus social), podemos extraer esa informacin antes de eliminar la columna. Pero en la mayora de los casos, especialmente si ests tratando con nombres reales de personas, es mejor eliminar la columna.

Las columnas `ID`, `SSN` y `Customer_ID` se eliminan generalmente por las siguientes razones:

**Irrelevancia para el modelado**: Estas columnas suelen ser identificadores nicos para cada registro o cliente. No contienen informacin til para el modelado predictivo, ya que no tienen ninguna relacin con las otras variables.

**Evitar el sobreajuste**: Si se incluyen en el modelo, pueden causar sobreajuste. El modelo podra aprender a hacer predicciones basndose en el ID del cliente, lo cual no es til ni generalizable.

**Privacidad de los datos**: En algunos casos, los identificadores de los clientes se eliminan para proteger la privacidad de los clientes, especialmente cuando los datos se comparten o se publican.

```python
# Eliminamos columnas que no aportan informacin relevante para el anlisis
data = data.drop(['ID', 'SSN', 'Customer_ID', 'Name', 'Month'], axis=1)
data.head()
```

#### Tratamiento de la columna `Age`

```python
# Mostramos los valores nicos de la columna 'Age'
print(list(data['Age'].unique()))
```

**Observacin:**
1. Algunos valores de `Age` tienen un guion bajo, que puede representar un error de tipeo (inconsistencia).
2. Se encuentran valores de edad negativos, lo que no es posible.
3. Aunque es posible que un cliente tenga ms de 100 aos, necesitamos establecer un rango.

```python
# Substituimos los valores '_' por '' y los convertimos a tipo flotante ('float')
data['Age'] = data['Age'].replace('_', '', regex=True).astype('float')

# Definimos un rango para los valores de la columna 'Age' de 1:100 y sustituimos los restantes con 'nan'
data.loc[data.Age < 1, 'Age'] = np.nan
data.loc[data.Age > 100, 'Age'] = np.nan

# Reemplazamos los valores 'nan' de la columna 'Age' con la media de la misma
data['Age'] = data['Age'].fillna(data['Age'].mean())

# Mostramos los valores nicos de la columna 'Age'
print(list(data['Age'].unique()))

# Mostramos el tipo de la columna 'Age'
print(data['Age'].dtype)
```

```python
# Revisamos que no haya valores faltantes y que el tipo de dato de la columna 'Age' sea correcto
data.info()
```

#### Tratamiento de la columna `Occupation`

```python
# Mostramos los valores nicos de la columna 'Occupation'
print(list(data['Occupation'].unique()))
```

**Observacin:**
1. Existe un valor '_______' que se debe eliminar.

```python
# Substituimos el valor '_______' por 'nan' y los convertimos a tipo flotante ('float')
data['Occupation'] = data['Occupation'].replace('_______', np.nan, regex=True)

# Mostramos el nmero e valores 'nan' en la columna 'Occupation'
print(data['Occupation'].isna().sum())
```

**Observacin:**
1. Tenemos un nmero considerable de registros `NaN`. Tenemos dos opciones: i) Reemplazas los valores `NaN` con la moda, no perders ningn registro. Pero en este contexto, estaramos imponiendo una profesin a clientes que posiblemente no llenaron ese campo. ii) La segunda opcin es eliminar los registros NaN, perders esos registros, pero tus datos sern ms precisos.

- Opcin 1: Reemplazar los datos con la moda:



```python
# Calculamos la moda
mode = data['Occupation'].mode()[0]
mode
```

```python
# Reemplazamos los valores NaN con la moda
data['Occupation'].fillna(mode, inplace=True)
```

- Opcin 2: Eliminar los registros NaN:

```python
# Eliminamos los registros con valores NaN en la columna 'Occupation'
data.dropna(subset=['Occupation'], inplace=True)
```

**Eliminar columnas con muchos valores faltantes**: Si una columna tiene una gran cantidad de valores faltantes (~>70%), podra ser ms til eliminarla que intentar imputar los valores faltantes.

En este ejemplo, no existen columnas con ms del 70% de valores faltantes. De existir, podemos eliminarla con el siguiente cdigo:

```python
# Eliminar columnas del DataFrame 'data' con ms del 70% de valores faltantes
threshold = int(len(data) * 0.7)
data = data.dropna(thresh=threshold, axis=1)
```

**Columnas con alta correlacin**: Si tienes dos columnas que estn altamente correlacionadas, podras considerar eliminar una de ellas. La razn es que ambas contienen informacin similar y tener ambas en el anlisis podra introducir multicolinealidad.



Recuerda que estas son solo pautas generales y la decisin final de qu columnas eliminar debe basarse en el conocimiento del dominio y el objetivo de tu anlisis.


**Codificacin de Variables Categricas**: Las variables categricas como 'Occupation' y 'Type_of_Loan' deben ser codificadas antes de ser usadas en un modelo. Podras usar codificacin one-hot o codificacin de etiquetas dependiendo de la naturaleza de los datos.


#### Identificacin y Manejo de Outliers

Los outliers pueden ser identificados utilizando mtodos estadsticos, como el mtodo del rango intercuartil (IQR). Una vez identificados, pueden ser eliminados o ajustados.



```python
# Identificar outliers con el mtodo IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Eliminar outliers
data = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]
```
