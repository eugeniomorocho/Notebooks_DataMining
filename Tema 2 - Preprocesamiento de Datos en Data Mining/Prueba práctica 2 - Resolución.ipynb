### **Fase I:** Carga y Preprocesamiento de Datos

#### Importamos las libreras necesarias

```python
# Importamos la librera 'numpy' con alias 'np'
import numpy as np

# Importamos la librera 'pandas' con alias 'pd'
import pandas as pd

# Importamos la librera 'matplotlib.pyplot' con alias 'plt'
import matplotlib.pyplot as plt

# Importamos la librera 'seaborn' con alias 'sns'
import seaborn as sns

# Importamos la librera 'datetime' con alias 'dt'
import datetime as dt

%matplotlib inline
```

#### Carga y preprocesamiento de datos

```python
# Leemos el archivo 'ecommerce_transactions.csv' y lo cargamos en un DataFrame con el nombre 'transacciones'
transacciones = pd.read_csv('Bases de Datos/ecommerce_transactions.csv')

# Mostramos las primeras 5 filas del DataFrame 'transacciones' usando el mtodo 'head()'
transacciones.head()
```

```python
# Revisamos los tipos de datos por cada columna con el mtodo 'info()' del DataFrame 'transacciones'
transacciones.info()
```

### **Fase II:** Limpieza de valores atpicos (outliers)

```python
# Revisamos los valores nulos en el DataFrame 'transacciones' con el mtodo 'isnull().sum()'
transacciones.isnull().sum()
```

```python
# Eliminamos valores nulos en el DataFrame 'transacciones' con el mtodo 'dropna()'
transacciones = transacciones.dropna()
```

```python
# Usamos el mtodo 'boxplot()' de la librera seaborn 'sns'
# Visualizamos la distribucin de la variable 'Quantity' del DataFrame 'transacciones'
sns.boxplot(y=transacciones['Quantity'])
plt.show()
```

```python
# Calculamos el primer cuartil (Q1) de la variable 'Quantity' del DataFrame 'transacciones'
# Usamos el mtodo 'quantile()' y el parmetro '0.25' 
Q1 = transacciones['Quantity'].quantile(0.25)

# Calculamos el segundo cuartil (Q2) de la variable 'Quantity' del DataFrame 'transacciones'
# Usamos el mtodo 'quantile()' y el parmetro '0.5'
Q2 = transacciones['Quantity'].quantile(0.75)

# Calculamos el rango intercuartil (IQR) de la variable 'Quantity' 
# restando el segundo cuartil (Q2) menos el primer cuartil (Q1)
IQR = Q2 - Q1
IQR
```

```python
# Definimos el lmite inferior usando la frmula Q1 - 1.5 * IQR
# Q1 es el primer cuartil, IQR es el rango intercuartil y 1.5 es un factor de escala
limite_inferior = Q1 - 1.5 * IQR

# Definimos el lmite superior usando la frmula Q2 + 1.5 * IQR 
# Q2 es el segundo cuartil, IQR es el rango intercuartil y 1.5 es un factor de escala
limite_superior = Q2 + 1.5 * IQR

# Filtramos los valores atpicos de la variable 'Quantity' del DataFrame 'transacciones' 
# Los outliers son aquellas observaciones que estn por debajo del lmite inferior o por encima del lmite superior
# Usamos el operador lgico '|' para indicar 'o' en la condicin
outliers = (transacciones['Quantity'] < limite_inferior) | (transacciones['Quantity'] > limite_superior)

# Eliminamos los valores atpicos de la variable 'Quantity' del DataFrame 'transacciones'
# Usamos el operador '~' para negar la condicin y obtener los valores que no son atpicos
transacciones = transacciones[~outliers]
```

```python
# Revisamos nuevamente los valores atpicos (outliers) en la columna 'Quantity'
# Usamos el mtodo 'boxplot()' de la librera 'seaborn' con la variable 'Quantity
sns.boxplot(y=transacciones['Quantity'])
```

```python
# Eliminamos los valores negativos de la columna 'Quantity' del DataFrame 'transacciones'
# Usamos el operador '>' para seleccionar los valores mayores a cero (0) en la columna 'Quantity'
transacciones = transacciones[transacciones['Quantity'] > 0]

# Revisamos nuevamente los valores atpicos (outliers) en la columna 'Quantity'
# Usamos el mtodo 'boxplot()' de la librera 'seaborn' con la variable 'Quantity' en el eje y
sns.boxplot(y=transacciones['Quantity'])
```

### **Fase III:** Productos que ms se venden y que mayor ganancia (revenue) generan

Creamos una columna nueva para las ganancias (revenue) por venta

```python
# Usamos el operador '*' para multiplicar las columnas 'Price' y 'Quantity' del DataFrame 'transacciones'
# Asignamos el resultado de la multiplicacin a la columna 'Revenue' del DataFrame 'transacciones'
transacciones['Revenue'] = transacciones['Price'] * transacciones['Quantity']

# Mostramos las primeras 5 filas del DataFrame 'transacciones' con el mtodo 'head()'
transacciones.head()
```

Revisamos los 10 productos que ms se han vendido

```python
# Agrupamos el DataFrame 'transacciones' por la columna 'ProductName'
# Usamos el mtodo 'groupby()' y pasamos la columna 'ProductName' como argumento
productos_agrupados = transacciones.groupby('ProductName')

# Sumamos la columna 'Quantity' de cada grupo
# Usamos el mtodo 'sum()' para sumar la columna 'Quantity' de cada grupo
suma_cantidades_por_producto = productos_agrupados['Quantity'].sum()

# Ordenamos los resultados de forma descendente
# Usamos el mtodo 'sort_values()' con el argumento 'ascending=False'
# Guardamos el resultado en la variable 'productos_mas_vendidos'
productos_mas_vendidos = suma_cantidades_por_producto.sort_values(ascending=False)

# Mostramos los 10 productos que ms se han vendido con el mtodo 'head()'
productos_mas_vendidos.head(10)
```

Visualizamos los 10 productos que ms se han vendido

```python
# Creamos una figura con el mtodo 'figure()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el tamao de la figura con el parmetro 'figsize' y las dimensiones (12, 6)
plt.figure(figsize=(12, 6))

# Usamos el mtodo 'barplot()' de la librera seaborn 'sns' para mostrar solo los 10 productos ms vendidos
# Definimos el eje x con los valores de la columna 'Quantity' de los 'productos_mas_vendidos'
# Definimos el eje y con los valores de los ndices de los 'productos_mas_vendidos'
# Definimos la paleta de colores con el parmetro 'palette' y el valor 'viridis'
sns.barplot(x=productos_mas_vendidos.head(10), y=productos_mas_vendidos.head(10).index, palette='viridis')

# Definimos el ttulo de la grfica con el mtodo 'title()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo 'Top 10 Productos ms Vendidos'
plt.title('Top 10 Productos ms Vendidos')

# Definimos la etiqueta del eje x con el mtodo 'xlabel()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo de la etiqueta del eje x como 'Cantidad'
plt.xlabel('Cantidad')

# Definimos la etiqueta del eje y con el mtodo 'ylabel()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo de la etiqueta del eje y como 'Producto'
plt.ylabel('Producto')

# Mostramos la grfica con el mtodo 'show()' de la librera matplotlib.pyplot con alias 'plt'
plt.show()
```

Revisamos los 10 productos que ms ingresos han generado

```python
# Agrupamos el DataFrame 'transacciones' por la columna 'ProductName' con el mtodo 'groupby()'
productos_agrupados = transacciones.groupby('ProductName')

# Sumamos los ingresos (Revenue) por cada producto usando el mtodo 'sum()'
suma_ingresos_por_producto = productos_agrupados['Revenue'].sum()

# Ordenamos los resultados de forma descendente
# Usamos el mtodo 'sort_values()' con el argumento 'ascending=False'
# Guardamos el resultado en la variable 'productos_mas_ingresos'
productos_mas_ingresos = suma_ingresos_por_producto.sort_values(ascending=False)

# Mostramos los 10 productos que ms ingresos han generado con el mtodo 'head()'
productos_mas_ingresos.head(10)
```

Visualizamos los productos que ms ingresos han generado

```python
# Creamos una figura con el mtodo 'figure()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el tamao de la figura con el parmetro 'figsize' y las dimensiones (12, 6)
plt.figure(figsize=(12, 6))

# Usamos el mtodo 'barplot()' de la librera seaborn 'sns' 
# Visualizamos solo los 10 productos que ms ingresos han generado 
# Definimos el eje x con los valores de la columna 'Revenue' que correspondan a los 'productos_mas_ingresos'
# Definimos el eje y con los valores de los ndices de los 'productos_mas_ingresos' 
sns.barplot(x=productos_mas_ingresos.head(10), y=productos_mas_ingresos.head(10).index, palette='viridis')

# Definimos el ttulo de la grfica con el mtodo 'title()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo 'Top 10 Productos con ms Ingresos'
plt.title('Top 10 Productos con ms Ingresos')

# Definimos la etiqueta del eje x con el mtodo 'xlabel()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo de la etiqueta del eje x como 'Ingresos'
plt.xlabel('Ingresos')

# Definimos la etiqueta del eje y con el mtodo 'ylabel()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo de la etiqueta del eje y como 'Producto'
plt.ylabel('Producto')

# Mostramos la grfica con el mtodo 'show()' de la librera matplotlib.pyplot con alias 'plt'
plt.show()
```

Visualizamos la distribucin de las ganancias por mes

```python
# Convertimos la columna 'Date' del DataFrame 'transacciones' a formato de fecha con el mtodo 'to_datetime()'
# Usamos el formato '%m/%d/%Y' para indicar el mes, da, y ao
# Guardamos el resultado en la columna 'Date' del DataFrame 'transacciones'
transacciones['Date'] = pd.to_datetime(transacciones['Date'], format  = '%m/%d/%Y')

# Creamos una nueva columna 'Month' en el DataFrame 'transacciones' con el mes de cada transaccin
# Usamos el mtodo 'dt.month' para obtener el mes de la columna 'Date'
transacciones['Month'] = transacciones['Date'].dt.month

# Agrupamos el DataFrame 'transacciones' por la columna 'Month' usando el mtodo 'groupby()'
# Sumamos los ingresos (Revenue) por mes usando el mtodo 'sum()'
# Guardamos los resultados en la variable 'ventas_por_mes'
ventas_por_mes = transacciones.groupby('Month')['Revenue'].sum()

# Creamos una figura con el mtodo 'figure()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el tamao de la figura con el parmetro 'figsize' y las dimensiones (12, 6)
plt.figure(figsize=(12, 6))

# Usamos el mtodo 'barplot()' de la librera seaborn 'sns' para visualizar los datos en forma de barras
# Definimos el eje x con los valores de los ndices de 'ventas_por_mes'
# Definimos el eje y con los valores de 'ventas_por_mes'
# Definimos la paleta de colores con el parmetro 'palette' y el valor 'viridis'
sns.barplot(x=ventas_por_mes.index, y=ventas_por_mes.values, palette='viridis')

# Definimos el ttulo de la grfica con el mtodo 'title()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo 'Ganancias por Mes'
plt.title('Ganancias por Mes')

# Definimos la etiqueta del eje x con el mtodo 'xlabel()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo de la etiqueta del eje x como 'Mes'
plt.xlabel('Mes')

# Definimos la etiqueta del eje y con el mtodo 'ylabel()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo de la etiqueta del eje y como 'Ganancias'
plt.ylabel('Ganancias')

# Mostramos la grfica con el mtodo 'show()' de la librera matplotlib.pyplot con alias 'plt'
plt.show()
```

Visualizamos la distribucin de las ventas por da de la semana con un grfico de barras

```python
# Visualizamos la distribucin de las ventas por da de la semana con un grfico de barras

# Creamos una nueva columna 'DayOfWeek en el DataFrame 'transacciones' con el da de la semana de cada transaccin
# Usamos el mtodo 'dt.day_name()' para obtener el nombre del da de la columna 'Date'
transacciones['DayOfWeek'] = transacciones['Date'].dt.day_name()

# Contamos el nmero de ventas por da de la semana con el mtodo 'value_counts()' de la columna 'DayOfWeek'
# Guardamos los resultados en la variable 'ventas_por_dia_semana'
ventas_por_dia_semana = transacciones['DayOfWeek'].value_counts()

# Creamos una figura con el mtodo 'figure()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el tamao de la figura con el parmetro 'figsize' y las dimensiones (12, 6)
plt.figure(figsize=(12, 6))

# Usamos el mtodo 'barplot()' de la librera seaborn 'sns' para visualizar los datos en forma de barras
# Definimos el eje x con los valores de los ndices de 'ventas_por_dia_semana'
# Definimos el eje y con los valores de 'ventas_por_dia_semana'
# Definimos la paleta de colores con el parmetro 'palette' y el valor 'viridis'
sns.barplot(x=ventas_por_dia_semana.index, y=ventas_por_dia_semana.values, palette='viridis')

# Definimos el ttulo de la grfica con el mtodo 'title()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo 'Ventas por Da de la Semana'
plt.title('Ventas por Da de la Semana')

# Definimos la etiqueta del eje x con el mtodo 'xlabel()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo de la etiqueta del eje x como 'Da de la Semana'
plt.xlabel('Da de la Semana')

# Definimos la etiqueta del eje y con el mtodo 'ylabel()' de la librera matplotlib.pyplot con alias 'plt'
# Definimos el ttulo de la etiqueta del eje y como 'Ventas'
plt.ylabel('Ventas')

# Mostramos la grfica con el mtodo 'show()' de la librera matplotlib.pyplot con alias 'plt'
plt.show()
```
